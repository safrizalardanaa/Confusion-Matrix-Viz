```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def conv_matrix_vis(con_matrix):
    # Add labels for group names and group counts
    group_names = ['True Negative', 'False Positive', 'False Negative', 'True Positive']
    group_counts = ["{0:0.0f}".format(value) for value in con_matrix.flatten()]

    # Calculate percentages for the confusion matrix
    group_percentages = ["{0:.2%}".format(value) for value in con_matrix.flatten() / np.sum(con_matrix)]

    # Combine group names, counts, and percentages
    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names, group_counts, group_percentages)]
    labels = np.asarray(labels).reshape(2, 2)

    # Create the heatmap
    plt.figure(figsize=(6, 4))
    sns.heatmap(con_matrix, annot=labels, fmt='', cmap='Blues', annot_kws={"fontsize": 12})

    # Set title
    plt.title('Confusion Matrix', fontsize=14)

    # Set axis labels
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)

    # Show the plot
    plt.show()

# Confusion matrix data
confusion_matrix_data = np.array([[592, 10],
                                  [43, 216]])

# Visualize the confusion matrix
conv_matrix_vis(confusion_matrix_data)
```

```python
def conv_matrix_vis(con_matrix):
    # Calculate class proportions
    class_proportions = con_matrix / np.sum(con_matrix)

    # Add labels for group names and group proportions
    group_names = ['True Negative', 'False Positive', 'False Negative', 'True Positive']
    group_counts = ["{0:0.0f}".format(value) for value in con_matrix.flatten()]

    # Calculate percentages for the confusion matrix
    group_proportions = ["{0:.2%}".format(value) for value in class_proportions.flatten()]

    # Combine group names and proportions
    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names, group_counts, group_proportions)]
    labels = np.asarray(labels).reshape(2, 2)

    # Set the range for annotation
    annotation_range = np.arange(0, 1.2, 0.2)

    # Create the heatmap using class proportions as values
    plt.figure(figsize=(6, 4))
    sns.heatmap(class_proportions, annot=labels, fmt='', cmap='Blues', annot_kws={"fontsize": 12},
                vmin=0, vmax=1, cbar_kws={'ticks': annotation_range})

    # Set title
    plt.title('Confusion Matrix', fontsize=12)

    # Set axis labels
    plt.xlabel('Predicted Label', fontsize=10)
    plt.ylabel('True Label', fontsize=10)

    # Show the plot
    plt.show()

# Confusion matrix data
convdata = np.array([[592,  10],
                     [ 43, 216]])

# Visualize the confusion matrix with proportions
conv_matrix_vis(convdata)
```

```python
def conv_matrix_vis(con_matrix):
    # Calculate class proportions
    class_proportions = con_matrix / np.sum(con_matrix)

    # Add labels for group names and group proportions
    group_counts = ["{0:0.0f}".format(value) for value in con_matrix.flatten()]
    group_proportions = ["{0:.2%}".format(value) for value in class_proportions.flatten()]

    # Combine group counts and proportions
    labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_counts, group_proportions)]
    labels = np.asarray(labels).reshape(con_matrix.shape[0], con_matrix.shape[1])

    # Set the range for annotation
    annotation_range = np.arange(0, 1.2, 0.2)

    # Define class labels
    class_labels = ["Class 0", "Class 1", "Class 2", "Class 3"]

    # Create the heatmap using class proportions as values
    plt.figure(figsize=(6, 4))
    sns.heatmap(class_proportions, annot=labels, fmt='', cmap='Blues', annot_kws={"fontsize": 11},
                vmin=0, vmax=1, cbar_kws={'ticks': annotation_range}, linewidths=.8, linecolor='blue',
                xticklabels=class_labels, yticklabels=class_labels)

    # Set title
    plt.title('Confusion Matrix', fontsize=14)

    # Set axis labels
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)

    # Show the plot
    plt.show()

# Confusion matrix data (example 4x4)
convdata = np.array([[592,  10, 30, 15],
                     [  3, 216,  5, 12],
                     [ 20,  8, 500, 40],
                     [ 10,  5,  12, 300]])

# Visualize the confusion matrix with proportions (4x4)
conv_matrix_vis(convdata)
```
